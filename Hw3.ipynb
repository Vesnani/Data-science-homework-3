```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Напишіть функцію гіпотези лінійної регресії у векторному вигляді

```python
def lin_regr(X, w):
  return np.dot(X, w)
```

Створіть функцію для обчислення функції втрат у векторному вигляді

```python
def loss_func(X, y, w):
  m = len(y)
  predictions = lin_regr(X, w)
  errors = predictions - y
  cost = (1 / (2 * m)) * np.dot(errors.T, errors)
  return cost
```

Реалізуйте один крок градієнтного спуску

```python
def gradient_step(X, y, w, learning_rate):
  m = len(y)
  predictions = lin_regr(X, w)
  errors = predictions - y
  gradient = (1 / m) * np.dot(X.T, errors)
  w = w - learning_rate * gradient
  return w
```

Знайдіть найкращі параметри $\vec{w}$ для датасету прогнозуючу ціну на будинок залежно від площі, кількості ванних кімнат та кількості спалень

```python
df = pd.read_csv('Housing.csv')
```

```python
X = df[['area', 'bedrooms', 'bathrooms']].values
y = df['price'].values
X = np.column_stack((np.ones(len(X)), X))
```

```python
mean_values = np.mean(X[:, 1:4], axis=0)
std_dev_values = np.std(X[:, 1:4], axis=0)

X[:, 1:4] = (X[:, 1:4] - mean_values) / std_dev_values
X = np.column_stack((np.ones(len(X)), X[:, 1:4]))

w_initial = np.zeros(X.shape[1])
learning_rate = 0.001
iterations = 100000

for params in range(iterations):
    w_initial = gradient_step(X, y, w_initial, learning_rate)

print("Найкращі параметри w:", w_initial)
```

Знайдіть ці ж параметри за допомогою аналітичного рішення

```python
w_analytical = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y)
print("Найкращі параметри w:", w_analytical)
```

Порівняйте отримані результати.

```python
loss_gradient = loss_func(X, y, w_initial)
loss_analytical = loss_func(X, y, w_analytical)

print("Функція втрат для градієнтного спуску:", loss_gradient)
print("Функція втрат для аналітичного рішення:", loss_analytical)
```
